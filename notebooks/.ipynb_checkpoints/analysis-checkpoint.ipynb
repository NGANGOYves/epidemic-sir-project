```python
# ===============================
# Epidemic Data Analysis
# COVID-19 Dataset
# ===============================

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Improve plot style
plt.style.use("ggplot")

# -------------------------------
# Load Dataset
# -------------------------------

data = pd.read_csv("../data/covid_19_data.csv")

print("Dataset loaded successfully")
print()

# Show first rows
print(data.head())

# -------------------------------
# Dataset Information
# -------------------------------

print("\nColumns in dataset:")
print(data.columns)

print("\nDataset shape:")
print(data.shape)

# -------------------------------
# Convert Date Column
# -------------------------------

data["ObservationDate"] = pd.to_datetime(data["ObservationDate"])

# -------------------------------
# Filter One Country
# -------------------------------

country = "Cameroon"

df = data[data["Country/Region"] == country]

print("\nFiltered data for:", country)
print(df.head())

# -------------------------------
# Aggregate Data by Date
# -------------------------------

df = df.groupby("ObservationDate").sum()

print("\nGrouped data:")
print(df.head())

# -------------------------------
# Plot Epidemic Curve
# -------------------------------

plt.figure(figsize=(10,6))

plt.plot(df.index, df["Confirmed"], label="Confirmed Cases")

plt.title("COVID-19 Epidemic Curve in Cameroon")
plt.xlabel("Date")
plt.ylabel("Number of Cases")

plt.legend()

plt.show()

# -------------------------------
# Save Figure for Report
# -------------------------------

plt.figure(figsize=(10,6))

plt.plot(df.index, df["Confirmed"], label="Confirmed Cases")

plt.title("COVID-19 Epidemic Curve in Cameroon")
plt.xlabel("Date")
plt.ylabel("Number of Cases")

plt.legend()

plt.savefig("../results/figures/epidemic_curve_cameroon.png")

plt.show()

print("\nFigure saved in results/figures/")
```
